```python
import geopandas as gpd
import rasterio
import matplotlib.pyplot as plt
```

```python
fwi_current = 'fwi_current.tif'

with rasterio.open(fwi_current) as src:
    array = src.read(1)
    
    transform = src.transform
    crs = src.crs

plt.figure(figsize=(10, 20))
plt.imshow(array, cmap='viridis',
           extent=(transform[2], transform[2]+transform[0]*array.shape[1], transform[5]+transform[4]* array.shape[0], transform[5]))
plt.title('FWI Map')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.colorbar(label='FWI Value')
plt.show()
```
